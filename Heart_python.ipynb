```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from math import e, pi, sin

def f(x, a): # definindo a função
   x = np.clip(x, -1.75, 1.75) # com x indo de -1,75 a 1,75
   return np.abs(x)**(2/3) + e/3 * np.maximum (0, (pi - x**2)**0.5) * np.sin(a * pi * x) # calculo da função


fig, ax = plt.subplots()
x = np.linspace(-2, 2, 300)
line, = ax.plot(x, f(x, 0), lw= 3, color='darkred')
ax.set_xlim([-2, 2])
ax.set_ylim([-1.5, 2.5])
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('f(x) = x^(2/3) + e/3*(π-x^2)^1/2*sin(a*π*x)')

def update(a):

  line.set_ydata(f(x, a))
  #ax.set_title(f'f(x) =  x^(2/3) + e/3*(π-x^2)^1/2*sin({a:.1f}πx)')
  return line,


# Criando a animação variando o valor de a de 1 até 20
ani = FuncAnimation(fig, update, frames=np.linspace(1, 40, 50), blit=True)

# Salvar a animação em um arquivo gif e exibir no Colab
ani.save('animation.gif', writer='pillow')

# Exibir a animação no Colab
from IPython.display import Image
Image(filename='animation.gif')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from math import e, pi, sin

# Definindo a função f(x, a)
def f(x, a):
   # Limita os valores de x no intervalo [-1.75, 1.75]
   x = np.clip(x, -1.75, 1.75)
   # Retorna o valor da função:
   # |x|^(2/3) + (e/3) * max(0, (π - x^2)^(1/2)) * sin(a * π * x)
   # Esta função combina uma exponenciação de x com uma componente senoide,
   # modulada pelo parâmetro 'a'.
   return np.abs(x)**(2/3) + e/3 * np.maximum(0, (pi - x**2)**0.5) * np.sin(a * pi * x)

# Cria a figura e o conjunto de eixos para o gráfico
fig, ax = plt.subplots()

# Gera 300 pontos igualmente espaçados no intervalo [-2, 2] para o eixo x
x = np.linspace(-2, 2, 300)

# Plota a função f(x, 0) (com a=0) e configura a linha do gráfico
# 'line,' representa o objeto da linha que será atualizado na animação
line, = ax.plot(x, f(x, 0), lw=3, color='darkred')

# Define os limites dos eixos x e y
ax.set_xlim([-2, 2])
ax.set_ylim([-1.5, 2.5])

# Define os rótulos dos eixos x e y
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

# Define o título do gráfico, exibindo a equação da função f(x)
ax.set_title('f(x) = x^(2/3) + e/3*(π-x^2)^1/2*sin(a*π*x)')

# Função para atualizar o gráfico durante a animação
def update(a):
   # Atualiza os dados do eixo y da linha com os novos valores de f(x, a)
   line.set_ydata(f(x, a))
   # Retorna a linha atualizada para o FuncAnimation
   return line,

# Cria a animação variando o valor de 'a' de 1 até 40 em 50 frames
ani = FuncAnimation(fig, update, frames=np.linspace(1, 40, 50), blit=True)

# Salva a animação como um arquivo GIF chamado 'animation.gif'
ani.save('animation.gif', writer='pillow')

# Exibe o GIF no ambiente de notebook (como Colab)
from IPython.display import Image
Image(filename='animation.gif')
```
